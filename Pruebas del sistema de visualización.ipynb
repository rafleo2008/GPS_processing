```python
import folium
import geopandas as gpd
proyecto = 'Peatonales'
geoZonesFilename = 'Cercas_ped.shp'
geoZonesFilenameP = proyecto +"/"+geoZonesFilename
print(geoZonesFilenameP)
geoZonesR = gpd.read_file(geoZonesFilenameP)

gpsFilename =  "Ped_cedritos.gpx"

gpsFilenameP = proyecto +"/"+gpsFilename
gpsPointsR = gpd.read_file(gpsFilenameP, layer = 'track_points')
```

```python
location = [4.723677,-74.031954]

m = folium.Map(location = location, zoom_start = 17)

## Dissolve all polygons and find centroid
geoZonesDiss = geoZonesR.dissolve()
geoJson = geoZonesDiss.to_json()
Geoel = folium.GeoJson(data = geoJson)
Geoel.add_to(m)

geoGPX = gpsPointsR.to_json()
gpxel = folium.GeoJson(data = geoGPX, popup='time')
gpxel.add_to(m)

## Project centroid to folium coord system


m


```

```python
select_widget=ipywidgets.Select(
    options=['Open Street Map','Terrain','Toner'],
    value='Open Street Map',
    description = 'Map Type',
    disabled=False)
def select(map_type):
    if map_type == 'Open Street Map':
        display(folium.Map(location=[4, -72], zoom_start=5, height=500))
    if map_type == 'Terrain':
        display(folium.Map(location=[4, -72], tiles =  'Stamen Terrain', zoom_start=5, height=500))
    if map_type == 'Toner':
        display(folium.Map(location=[4, -72], tiles =  'Stamen Toner', zoom_start=5, height=500))
        
ipywidgets.interact(select,map_type=select_widget)
```
